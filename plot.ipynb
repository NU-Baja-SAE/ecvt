```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
LOW_GEAR = 3.6
HIGH_GEAR = 0.9

SLIP_SPEED = 600 # RPM
CRUISE_LOW = 3000 / LOW_GEAR
CRUISE_HIGH = 3000 / HIGH_GEAR



def gear_ratio_from_secondary_rpm(x):
    if (x < SLIP_SPEED):
        return LOW_GEAR # maybe somewhere between LOW and IDLE
    elif(x < CRUISE_LOW):
        return LOW_GEAR
    elif(x < CRUISE_HIGH):
        return (LOW_GEAR - HIGH_GEAR) / (CRUISE_LOW - CRUISE_HIGH) * (x - CRUISE_LOW) + LOW_GEAR
    else:
        return HIGH_GEAR
```

```python
x = np.linspace(0,4000)

y = [gear_ratio_from_secondary_rpm(i) for i in x]
plt.plot(x,y)
plt.xlabel("Secondary RPM")
plt.ylabel("Gear Ratio")
plt.title("Gear Ratio vs Secondary RPM")
plt.show()
```
